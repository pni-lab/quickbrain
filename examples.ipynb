# quickbrain — examples

**quickbrain** provides fast, lightweight whole-brain surface plotting with a single function call.
Unlike most neuroimaging visualisation packages, the built-in MNI152 meshes include the
**cerebellum and brainstem**, giving you a complete overview of brain-wide statistical maps.

```python
from quickbrain import plot_brain
```

That's the only import you need.

````python
from quickbrain import plot_brain
````

## 3. Plotting a NIfTI volume

Pass any NIfTI path (or a loaded nibabel image) as `stat_map`.
quickbrain projects the volume onto the surface automatically via
`nilearn.surface.vol_to_surf`.

Here we use the Pain-activation stat map bundled with nilearn.

````python

from matplotlib import pyplot as plt


fig, axes = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={"projection": "3d"})


plot_brain(pain_response,
           threshold=0, 
           res=10,
           title="Pain response (Hedges' g)",
           post_blur=3,
           background="transparent",
           vol_to_surf_radius=3,
           overlay_linewidth=1,
           colorbar=False,
           cmap='coolwarm', #"cold_hot",
           axes=axes[0],
           vmin=-0.5,
           vmax=0.5,
           );

plot_brain(pain_response,
           threshold=0, 
           res=1,
           title="Pain response (Hedges' g)",
           post_blur=3,
           background="transparent",
           vol_to_surf_radius=3,
           overlay_linewidth=1,
           colorbar=False,
           cmap='coolwarm', #"cold_hot",
           axes=axes[1],
           vmin=-0.5,
           vmax=0.5,
           );
````

## 5. Resolution levels

The `res` parameter controls mesh density (voxel sizes 1–10 used during
simplification).  Lower `res` = finer mesh.  `res=3` (default) is a
good balance of speed and detail.

````python
fig, (ax_p, ax_i) = plt.subplots(
    1, 2,
    figsize=(12, 5),
    subplot_kw={"projection": "3d"},
)

for ax, st in [(ax_p, "pial"), (ax_i, "inflated")]:
    plot_brain(
        pain_response,
        surf_type=st,
        threshold=0.5,
        axes=ax,
        colorbar=False,
        title=st,
        overlay=None,
    )

fig.suptitle("Pial vs. inflated", fontsize=13)
plt.tight_layout()
plt.show()
````

## 7. Background options

The `bg_map` parameter controls the underlay:

| value | effect |
|---|---|
| `"curvature"` (default) | signed curvature — highlights gyri/sulci |
| a float, e.g. `0.5` | uniform grey |
| `None` | no background (white) |

````python
fig, (ax1, ax2, ax3) = plt.subplots(
    1, 3,
    figsize=(15, 5),
    subplot_kw={"projection": "3d"},
)

for ax, bg, label in [
    (ax1, "curvature", 'bg_map="curvature"'),
    (ax2, 0.5, "bg_map=0.5"),
    (ax3, 0.1, "bg_map=None"),
]:
    plot_brain(
        pain_response,
        bg_map=bg,
        threshold=0.5,
        axes=ax,
        colorbar=False,
        title=label,
        overlay=None,
    )

plt.tight_layout()
plt.show()
````

## 10. Saving to file

Use `output_file` to write the figure directly (any format matplotlib

## 11. Full four-panel overview

A common use-case: show both hemispheres in lateral and medial views
in a single figure.  Four lines of `plot_brain` and a shared
`suptitle` — that's it.

````python
fig, axes = plt.subplots(
    2, 2,
    figsize=(12, 10),
    subplot_kw={"projection": "3d"},
)

for i, hemi in enumerate(["left", "right"]):
    for j, view in enumerate(["lateral", "medial"]):
        plot_brain(
            pain_response,
            hemi=hemi,
            view=view,
            threshold=0.5,
            axes=axes[i, j],
            colorbar=(j == 1)
        )

fig.suptitle(
    "Pain activation — full overview (incl. cerebellum)",
    fontsize=14,
    y=1.01,
)
plt.tight_layout()
plt.show()
````

## 13. Post-processing blur

`post_blur` (default 3 px) applies a light Gaussian blur to the brain
rendering *before* the SVG overlay is composited.  This softens the
polygon edges.  Set to `0` to disable.

## 14. Background: white vs. transparent

`background="transparent"` turns the white canvas transparent (useful for
compositing into posters or slides).  Only two values are allowed:
`"white"` (default) and `"transparent"`.

## 15. Overlay line width

`overlay_linewidth` is a multiplier for the SVG stroke widths.
Use values > 1 for thicker contours, < 1 for thinner ones.
